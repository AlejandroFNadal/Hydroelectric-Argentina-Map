# Making a map of renewable power stations in Argentina

# Folium : Interactive Map

```python
import folium
from folium import Marker
from folium.plugins import MarkerCluster, FloatImage
import math
```

```python
map_renewables = folium.Map(location=[-35,-58], tiles='openstreetmap', zoom_start=4)
```

```python
map_type_color=[['BIOMASA','#FF0000'],
 ['EÓLICA','#2F4F2F'],
 ['SOLAR', '#FFFF00'],
 ['PAH','#FF7F00'],
 ['BIOGÁS','#800080'],
 ['BRS','#000000']]
def get_color(energy_type:str):
    for pair in map_type_color:
        if energy_type == pair[0]:
            return pair[1]
    raise ValueError('Not valid energy type')
```

```python
#to generate partial circles to show non finished plants
def polarToCartesian(centerx, centery, radius, angle):
    pos_neg = 90-angle
    if pos_neg < 0:
        non_neg = 360 + pos_neg
    else:
        non_neg = pos_neg
    angle_radian = (non_neg) * math.pi / 180
    x = centerx + (radius * math.cos(angle_radian))
    y = centery - (radius * math.sin(angle_radian))
    return x,y

def describeArc(x,y,radius,start_angle, end_angle):
    start = polarToCartesian(x,y,radius,start_angle)
    end = polarToCartesian(x,y,radius,end_angle)
    if end_angle >=180:
        big_arc = "1"
    else:
        big_arc = "0"
    large_arc_flag = "1" 
   
    d = f"M {x} {y} L {x} {y-radius} A {radius} {radius} 0 {big_arc}, {large_arc_flag}, {end[0]},{end[1]} z"
    return d
```

```python

#now we add the markers
mc = MarkerCluster()
for i in range(0, len(argentina)):
    row = argentina.iloc[i]
    color = get_color(row['tecnologia'])
    avance = float(row['avance'])
    if avance == 1:
        folium.Marker(
        location = [argentina.iloc[i]['latitud'], argentina.iloc[i]['longitud']],
                    popup=folium.Popup(f"Nombre: {argentina.iloc[i]['proyecto']}\n Potencia: {argentina.iloc[i]['potencia_mw']}", parse_html=True),
                    icon=folium.DivIcon(html=f"""<div><svg><circle cx="20", cy="20", r="15" fill="{color}" stroke="black" stroke-width="1"/></svg></div>""")
                    ).add_to(mc)
    elif avance > 0 and avance < 1:
        # if avance is lower than .25 it becomes really hard to see
        if avance < 0.25:
            avance = 0.25
        path = describeArc(20,20,15,0,avance*360)
        folium.Marker(
            location = [argentina.iloc[i]['latitud'], argentina.iloc[i]['longitud']],
                        popup=folium.Popup(f"Nombre: {argentina.iloc[i]['proyecto']}\n Potencia: {argentina.iloc[i]['potencia_mw']}", parse_html=True),
                        icon=folium.DivIcon(html=f"""<div style="text-align:center;"><svg><path d="{path}" stroke="black"  fill="{color}" /></svg></div>""")
                        ).add_to(mc)
map_renewables.add_child(mc) 
```

```python
import base64
```

```python
with open('./images/legend.jpg', 'rb') as lf:
  # open in binary mode, read bytes, encode, decode obtained bytes as utf-8 string
    b64_content = base64.b64encode(lf.read()).decode('utf-8')
    FloatImage('data:image/png;base64,{}'.format(b64_content), bottom=5, left=70).add_to(map_renewables)
```
